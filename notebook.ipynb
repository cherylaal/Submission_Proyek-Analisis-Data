### Cleaning Data

#### Cleaning Data geolocation_df

```python
# Menangani duplikasi data dengan menghapus duplikasi data berdasarkan kolom geolocation_zip_code_prefix dan mempertahankan nilai pertama
geolocation_df.drop_duplicates(
    subset=["geolocation_zip_code_prefix"], keep="first", inplace=True
)
```

```python
#Memeriksa adanya duplikasi data
print("Jumlah duplikasi pada data geolocation_df: ", geolocation_df.duplicated().sum())
```

#### Cleaning Data order_items_df

```python
#Menangani outlier pada kolom price menggunakan IQR

Q1 = order_items_df["price"].quantile(0.25)
Q3 = order_items_df["price"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

order_items_df = order_items_df[
    (order_items_df["price"] >= lower_bound) & (order_items_df["price"] <= upper_bound)
]
```

```python
#Memeriksa kembali keberadaan outlier pada data yang sudah ditangani
plt.figure(figsize=(12, 6))
sns.boxplot(x=order_items_df["price"])
plt.title("Boxplot of Price after Removing Outliers")
plt.show()
```

```python
#Menangani outlier pada kolom freight_value menggunakan IQR

Q1 = order_items_df["freight_value"].quantile(0.25)
Q3 = order_items_df["freight_value"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

order_items_df = order_items_df[
    (order_items_df["freight_value"] >= lower_bound)
    & (order_items_df["freight_value"] <= upper_bound)
]
```

```python
#Memeriksa kembali keberadaan outlier pada data yang sudah ditangani
plt.figure(figsize=(12, 6))
sns.boxplot(x=order_items_df["freight_value"])
plt.title("Boxplot of Freight Value after Removing Outliers")
plt.show()
```

```python
order_items_df.describe()
```

#### Cleaning Data orders_df

```python
datetime_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date','order_delivered_customer_date', 'order_estimated_delivery_date']

#Mengubah tipe data datetime
for col in datetime_cols:
    orders_df[col] = pd.to_datetime(orders_df[col])
```

```python
#Memeriksa tipe data yang telah diubah
orders_df.info()
```

```python
# Mengurutkan waktu berdasarkan kolom order_purchase_timestamp sebelum menangani missing value
orders_df.sort_values(by=["order_purchase_timestamp"], inplace=True)
```

```python
orders_df.head(10)
```

```python
#Menghapus 3 baris teratas karena terdapat 3 missing value untuk menghndari ketidak-akurat an data ketika dilakukan penanganan menggunakan forward fill
orders_df.drop(orders_df.index[:3], inplace=True)
```

```python
orders_df.head()
```

```python
# Mengganti missing value dengan nilai sebelumnya (forward fill)
missing_datetime_cols = ['order_approved_at', 'order_delivered_carrier_date','order_delivered_customer_date']

for col in missing_datetime_cols:
  orders_df[col] = orders_df[col].fillna(method='ffill')
```

```python
#Memeriksa jumlah missing value yang sudah ditangani
orders_df.isna().sum()
```

#### Cleaning Data products_df

```python
#Mengatasi missing value kolom kategorikal
#Mengganti missing value dengan nilai modus/mode dari product_category_name
mode_product_category_name = products_df["product_category_name"].mode()[0]
products_df["product_category_name"].fillna(mode_product_category_name, inplace=True)
```

```python
#Memeriksa jumlah missing value
products_df.isna().sum()
```

```python
#Mengatasi missing value kolom numerik
#Mengganti missing value dengan nilai median dari masing masing kolom
numeric_cols_product = ['product_name_lenght','product_description_lenght','product_photos_qty','product_weight_g','product_length_cm','product_height_cm','product_width_cm']

for col in numeric_cols_product:
    products_df[col] = products_df[col].fillna(products_df[col].median())
```

```python
#Memeriksa jumlah missing value
products_df.isna().sum()
```

## Exploratory Data Analysis (EDA)

### Explore orders_df

Jumlah pembelian dalam hari

```python
orders_df["Day"] = orders_df["order_purchase_timestamp"].dt.dayofweek.map({
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
})

# Buat kolom baru "Day"
orders_df["Day"] = orders_df["order_purchase_timestamp"].dt.strftime("%d-%m-%Y %H:%M:%S")
orders_df["Day"] = orders_df["Day"].apply(lambda x: datetime.strptime(x, "%d-%m-%Y %H:%M:%S").strftime("%A"))

# Menghitung jumlah pembelian per hari
df_day = orders_df.groupby("Day").size().reset_index(name="purchase_amount")

# Temukan hari dengan pembelian terbanyak
day_favorit = df_day["Day"].loc[df_day["purchase_amount"] == df_day["purchase_amount"].max()]

print(f"Hari favorit untuk berbelanja: {day_favorit}")
df_day
```

Jumlah pembelian dalam jam

```python
# Buat kolom baru "hours"
orders_df["hours"] = orders_df["order_purchase_timestamp"].dt.hour

# Menghitung jumlah pembelian per jam
df_hours = orders_df.groupby("hours").size().reset_index(name="purchase_amount")

# Temukan jam dengan pembelian terbanyak
hours_favorit = df_hours["hours"].loc[df_hours["purchase_amount"] == df_hours["purchase_amount"].max()]

print(f"Jam favorit untuk berbelanja: {hours_favorit}")
df_hours
```

### Explore orders_df, order_items_df, dan products_df

Pada data orders_df dan order_items_df terdapat kolom yang berisi informasi terkait penjualan produk. Informasi ini dapat digunakan untuk mengidentifikasi perkembangan penjualan produk pada periode tertentu.

```python
#Menggabungkan orders_df dengan order_items_df menggunakan order_id
sales_df = orders_df.merge(order_items_df, on="order_id")
sales_df.head()
```

```python
#Menggabungkan orders_df, order_items_df, dan products_df menggunakan product_id
sales_df = sales_df.merge(products_df, on="product_id")
sales_df.head()
```

```python
# Menambahkan kolom tahun
sales_df['year'] = sales_df['order_purchase_timestamp'].dt.year
sales_df.head()
```

```python
sales_df.info()
```

```python
#Melihat performa penjualan berdasarkan kategori produk dan tahun penjualan
grouped_sales_df = sales_df.groupby(by=["product_category_name","year"]).agg({
    "order_id": "nunique",
    "order_item_id": "count",
    "price": "sum"
})
#Mengganti nama kolom order_item_id menjadi quantity
grouped_sales_df = grouped_sales_df.rename(columns={"order_item_id": "quantity"})
grouped_sales_df.head(15)
```

#### Perkembangan penjualan berdasarkan harga

```python
#Group data berdasarkan product_category_name dan year
sales_grouped_price_df = grouped_sales_df.groupby(['product_category_name', 'year'])['price'].sum()

# Membuat dataframe dengan tahun sebagai kolom
sales_grouped_price_df = sales_grouped_price_df.unstack()
sales_grouped_price_df.head(10)
```

```python
#Menghitung pertumbuhan YoY (Year-over-Year) untuk setiap kategori produk
for col in sales_grouped_price_df.columns:
  sales_grouped_price_df[f'{col}_YoY_Growth'] = sales_grouped_price_df[col].diff()

sales_grouped_price_df = sales_grouped_price_df.iloc[:, 4:]
sales_grouped_price_df.head(10)
```

```python
#Mencari kategori dengan peningkatan rata-rata tertinggi
category_growth_price_df = sales_grouped_price_df.mean(axis=1)

print(category_growth_price_df)
```

```python
#Mengurutkan hasil rata-rata
highest_growth_price_df = category_growth_price_df.sort_values(ascending=False)

#Menampilkan 10 kategori produk dengan rata rata peningkatan tertinggi berdasarkan harga
top_10_highest_growth_price_categories = highest_growth_price_df.head(10).index

print("10 Kategori dengan Pertumbuhan Penjualan Tertinggi Berdasarkan Harga(2016-2018):")
for category in top_10_highest_growth_price_categories:
  category_growth_price = highest_growth_price_df.loc[category].round(2)
  print(f"- {category}: {category_growth_price}")
```

#### Perkembangan penjualan berdasarkan quantity

```python
#Group data berdasarkan product_category_name dan year
sales_grouped_qty_df = grouped_sales_df.groupby(['product_category_name', 'year'])['quantity'].sum()

#Membuat dataframe dengan tahun sebagai kolom
sales_grouped_qty_df = sales_grouped_qty_df.unstack()
sales_grouped_qty_df.head(10)
```

```python
#Menghitung pertumbuhan YoY (Year-over-Year) untuk setiap kategori produk
for col in sales_grouped_qty_df.columns:

  sales_grouped_qty_df[f'{col}_YoY_Growth'] = sales_grouped_qty_df[col].diff()

sales_grouped_qty_df = sales_grouped_qty_df.iloc[:, 4:]
sales_grouped_qty_df.head(10)
```

```python
# Mencari kategori dengan pertumbuhan YoY rata-rata tertinggi
category_growth_qty_df = sales_grouped_qty_df.mean(axis=1)

category_growth_qty_df
```

```python
#Mencari kategori dengan peningkatan rata-rata tertinggi#Mengurutkan hasil rata-rata
highest_growth_qty_df = category_growth_qty_df.sort_values(ascending=False)

#Menampilkan 10 kategori produk dengan rata rata peningkatan tertinggi berdasarkan quantity
top_10_highest_growth_qty_categories = highest_growth_qty_df.head(10).index

print("10 Kategori dengan Pertumbuhan Penjualan Tertinggi Berdasarkan Quantity(2016-2018):")
for category in top_10_highest_growth_qty_categories:
  category_growth_qty = highest_growth_qty_df.loc[category].round(2)
  print(f"- {category}: {category_growth_qty}")
```
